```python
import numpy as np
class_1 = len(newdf['Stages'][newdf.Stages == 0])
class_2 = len(newdf['Stages'][newdf.Stages == 1])
class_3 = len(newdf['Stages'][newdf.Stages == 2])

labels = ['Class_1', "Class_2", "Class_3"]

arr = np.array([class_1, class_2, class_3])

plt.pie(arr, labels=labels, explode=[0.0, 0.2, 0.2], shadow=True)
plt.legend()
```

```python
from sklearn.model_selection import train_test_split

x = newdf.drop('Stages', axis=1)
y = newdf['Stages']

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=2)

print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression().fit(x_train, y_train)

lr_pred = lr.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, recall_score, precision_score, f1_score

def model_evaluation(y_test, y_pred):
    print("Accuracy Score:", accuracy_score(y_test, y_pred))
```

```python
model_evaluation(y_test, lr_pred)
```

```python
from sklearn.naive_bayes import MultinomialNB
model = MultinomialNB()
model.fit(x_train, y_train)
```

```python
nb_pred = model.predict(x_test)
```

```python
model_evaluation(y_test, nb_pred)
```

```python
from sklearn.svm import SVC
svc = SVC().fit(x_train, y_train)
svc_pred = svc.predict(x_test)
```

```python
model_evaluation(y_test, svc_pred)
```
